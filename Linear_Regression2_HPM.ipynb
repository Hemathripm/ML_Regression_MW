```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('dataset_Facebook.csv')
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
x=data.drop('Total Interactions',axis=1)
```

```python
y=data[['Total Interactions']]
```

```python
x=pd.get_dummies(x)
```

```python
#Standard Scalar Transformation
from sklearn.preprocessing import StandardScaler
x_scaled=StandardScaler()
x_scaled.fit(x)
x_scaled=x_scaled.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.4,random_state=1)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```
